```python
# 라이브러리 불러오기 및 별명 지정하기
import pandas as pd
import matplotlib.pyplot as plt
# 한글폰트 적용하기
plt.rc('font', family='Malgun Gothic')

# csv 데이터 불러오기
lotte = pd.read_csv("롯데월드 CSV.csv", encoding='cp949', thousands=',')
lotte
```

```python
# 불필요한 컬럼 삭제
del lotte['시도']
del lotte['군구']
del lotte['관광지']
del lotte['내/외국인']
del lotte['총계(2018.01 ~ 2023.12)']
del lotte['인원계']
del lotte['인원계.1']
del lotte['인원계.2']
del lotte['인원계.3']
del lotte['인원계.4']
del lotte['인원계.5']
del lotte['Unnamed: 83']

lotte.loc[0]
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(12,5))
plt.title('롯데월드 입장객 통계(2018년~2023년)')

lotte.loc[0][0:12].plot(label='2018년', marker='o')
lotte.loc[0][13:25].plot(label='2019년', marker='o')
lotte.loc[0][25:37].plot(label='2020년', marker='o')
lotte.loc[0][37:49].plot(label='2021년', marker='o')
lotte.loc[0][49:61].plot(label='2022년', marker='o')
lotte.loc[0][60:73].plot(label='2023년', marker='o')

plt.legend()
```
